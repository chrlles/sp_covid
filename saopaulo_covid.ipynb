## São Paulo: Casos e óbitos  por município 

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dados = pd.read_csv("_dados_covid_municipios_sp.csv", sep=';', encoding='latin1')
dados.head()
```

```python
dados = dados.rename(columns={'nome_munic':'Município'})
```

```python
dados.tail()
```

```python
dados.info()
```

```python
dados = dados.dropna()
```

```python
dados.describe()
```

```python
dados['datahora'] = pd.to_datetime(dados['datahora'], format = '%d/%m/%Y')
```

```python
dados.head()
```

```python
dados.shape
```

```python
# Média casos COVID-19 nas Regiões Administrrativas
dados.groupby('nome_ra').agg({'casos':'mean'})
```

```python
# Municipios e os caso de COVID-19
dados.groupby('Município').agg({'casos':'sum'}).sort_values(by='casos', ascending=False)
```

```python
# Total de casos
sum(dados['casos'])
```

```python
# População total - SP
sum(dados['pop'])
```

```python
# máximo de obitos em um dia
max(dados['obitos'])
```

```python
lilas = '#8E72A3'
```

```python
fig, ax = plt.subplots(figsize=(12,5))
sns.lineplot(data=dados, x='datahora', y='casos_mm7d', ci=None, color=lilas)
plt.xlabel('data', fontsize=12)
plt.title('Média móvel dos últimos 7 dias dos novos casos', fontsize=12)
plt.ylabel('Média movel de casos últimos 7 dias', fontsize=12)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
```

```python
verde = '#43734F'
```

```python
fig, ax = plt.subplots(figsize=(16,7))
sns.lineplot(data=dados, x='datahora', y='casos_mm7d', ci=None, color=verde, ms=3)
plt.xlabel('data', fontsize=15)
plt.title('Média móvel dos últimos 7 dias dos novos casos', fontsize=15)
plt.ylabel('Média movel de casos últimos 7 dias', fontsize=14)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
```

```python
fig, ax = plt.subplots(figsize=(16,7))
sns.lineplot(data=dados, x='datahora', y='obitos_mm7d', color='dimgrey', ci=None)
plt.title('Média móvel dos últimos 7 dias dos novos obitos', fontsize=15)
plt.xlabel('data', fontsize=15)
plt.ylabel('Média movel óbitos últimos 7 dias', fontsize=15)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
```

```python
# Gráfico de letalidade durante as semanas de pandemia, um crescimento entre a 14 semana e 17;
#  Queda no número de letalidade de COVID-19 entre a 25 semana e a 36 semana.
plt.figure(figsize=(15,5))
sns.barplot(data=dados, x='semana_epidem', y='letalidade', ci=None, color="#6663AB")
plt.title('letalidade X semana de epidemia', fontsize=18)
plt.xlabel('semana de epidemia', fontsize=15)
plt.ylabel('letalidade', fontsize=15)
```
